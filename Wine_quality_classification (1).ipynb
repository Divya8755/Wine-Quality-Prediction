#Importing Libearies


```python
import numpy as np
import pandas as pd
```

#Loading the data

```python
dataset=pd.read_csv('/content/WineQT.csv')
```

#Checking for null values

```python
dataset.isnull().sum()
```

#Finding correlation


```python
def corr_features(corr_df,threshold):
  arr=corr_df.to_numpy()
  feat=set()
  for i in range(len(arr)):
    for j in range(len(arr)):
      if(j>i):
        if(arr[i,j]>=threshold or arr[i,j]<=-threshold):
         feat.add(corr_df.columns[j])
  return feat
```

```python
corr_df=dataset.corr(method="spearman")
feat=corr_features(corr_df,0.60)
print(feat)
```

#Dropping correlated values

```python
dataset=dataset.drop('total sulfur dioxide',axis=1)
dataset=dataset.drop('pH',axis=1)
dataset=dataset.drop('citric acid',axis=1)
dataset=dataset.drop('density',axis=1)
dataset.shape
```

```python
dataset.shape
```

#Classifying the quality of wine in good(1) or bad(0)

```python

dataset['quality2'] = [1 if x >= 5 else 0 for x in dataset['quality']]
dataset=dataset.drop('quality',axis=1)
```

#Dropping unnecssary features

```python
dataset.drop('Id',axis=1)
```

```python
dataset.head()
```

#Splitting data according to dependent and independent featurex

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
m,n=X.shape
```

#Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

#Training the  model on the Training set

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 500, criterion = 'entropy', random_state = 0,n_jobs=-1)
classifier.fit(X_train, y_train)
```

#Saving the model

```python
import joblib
#saving the model to the Disk
file_name='Wine_quality_model'
joblib.dump(classifier,file_name)
```

#Loading the model from disk

```python
import joblib
#loading the model from Disk
loaded_model=joblib.load('/content/Wine_quality_model')
```

#Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

#Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
#confusion matrix
print(cm)
#calculating the accuracy score of the model
print("accuracy")
print((accuracy_score(y_test, y_pred)*100))



```
